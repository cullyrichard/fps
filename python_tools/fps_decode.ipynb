```python
def getcode(val):
    global opcodes
    if opcodes is not None:
        op=opcodes.get(val.pop(0),{})
        while len(val)>0:
            if 'codes' in op:
                op=op['codes']
            op=op.get(val.pop(0),{})
        if 'mn' in op:
            return f"({op['mn']})"
        elif 'name' in op:
            return f"({op['name']})"
    return ''
```

```python
def parce_code(opcat):
    global opcodes
    inst=[]
    value=False
    disabled=set()
    for code in opcodes.values():
        op=code
        if code.get('name') in disabled:
            continue
        mask,offset=op.get('range')
        val=(opcat>>offset)&mask
        while 'codes' in op:
            mask,offset=op.get('range')
            val=(opcat>>offset)&mask
            #print(op.get('name'),63-offset,mask,val)
            disabled.update(op.get('disable',[]))
            codes=op.get('codes',{})
            if val in codes:
                op=codes.get(val)
            else:
                break
        value|='value' in op
        disabled.update(op.get('disable',[]))
        if offset<16 and value:
            break
        mn=op.get('mn')
        name=op.get('name','')
        if mn:
            if 'IDX' in op:
                mask,offset=opcodes.get(op.get('IDX')).get('range')
                val=(opcat>>offset)&mask
                mn=mn.replace('IDX',f'{val-4}')
            inst.append(mn)
        if 'data' in op:
            inst.append(f'0{val:o}')
        #print(name,mn,63-offset,bin(mask),(opcat>>offset)&mask,value,disabled,'data' in op)
    #print(disabled)
    sinst=";".join(inst)
    if value:
        if 'VALUE' in sinst:
            sinst=sinst.replace('VALUE',f'0{op3:o}')
        else:
            sinst+=f' VAL: 0{op3:o}'
    return sinst
```

```python
import re
def get_asm_part(part):
    loc=part.get('range')
    codes={c.get('mn'):{'nm':c.get('mn'),'val':v,'range':loc} for v,c in part.get('codes').items() if 'mn' in c}
    rex='|'.join(['\\b'+c+'\\b' for c in codes.keys()])
    return rex,codes
```

```python
def set_opcat_data(path,opcat,value):
    global opcodes
    op=opcodes
    for p in path:
        op=op.get(p)
    mask,offset=op.get('range')
    opcat|=(value&mask)<<offset
    return opcat
```

```python
opcode_lookup={}

def process_codes(op,deps,tree):
    codes=op.get('codes')
    for code,data in codes.items():
        print(code,data)
        item=data.copy()
        print(op.get('range',()))
        dep=deps.copy()
        range_val=op.get('range',[]).copy()
        range_val.append(code)
        dep.append(range_val)
        print(dep)
        if 'mn' in item:
            item['dep']=dep
            opcode_lookup[tree][item.get('mn')]=item
        elif 'name' in item:
            name=item.get('name')
            opcode_lookup[name]={}
            process_codes(data,dep,item.get('name'))


for opcode in opcodes.values():
    if 'codes' in opcode:
        print(opcode)
        name=opcode.get('name')
        opcode_lookup[name]={}
        process_codes(opcode,[],name)

```
